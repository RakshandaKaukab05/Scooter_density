```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import folium

%matplotlib inline
```

```python
july_df = pd.DataFrame()
for chunk in pd.read_csv('./data/july.csv', chunksize = 1000000, low_memory=False):
    july_df = pd.concat([july_df,chunk])
```

```python
july_df.sumdgroup.value_counts()
```

```python
to_drop = july_df[july_df['sumdgroup']=='bicycle']
july_df = july_df.drop(to_drop.index, axis = 0)
july_df.head()
```

```python
july_df.companyname.value_counts()
```

```python
july_new = july_df["pubdatetime"].str.split(" ", n = 1, expand = True)
july_df["date"]= july_new[0]
july_df["time"]= july_new[1]
july_new.head()
```

```python
july_df.head()
```

```python
july_df  = july_df.drop (columns =['pubdatetime'])
july_df.head()
```

```python
july_df['date']=pd.to_datetime(july_df['date'],format='%Y-%m-%d')
```

```python
july_df['time']=pd.to_datetime(july_df.time)
```

```python
july_df.sumdid.head(5)
```

```python
july_df = july_df[july_df['sumdid'].isin(['Powered570380'])]
july_df.head()
```

```python
july_df['Time Elapsed'] = july_df.sort_values(['date', 'time']).groupby('date')['time'].diff()
july_df.head(10)
```

```python
scooter_map = folium.Map(location = [36.1612, -86.7775], zoom_start = 12)


for ind, row in july_df.iterrows():
    point = [row['latitude'], row['longitude']]
    popup = (row['time'],row['Time Elapsed'])
    folium.Marker(point, popup).add_to(scooter_map)

scooter_map
```
